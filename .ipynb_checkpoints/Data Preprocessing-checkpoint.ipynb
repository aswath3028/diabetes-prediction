```python
#Outliers removal

Q1=data.quantile(0.25)
Q3=data.quantile(0.75)
IQR=Q3-Q1

print("---dQ1--- \n",Q1)
print("\n---Q3--- \n",Q3)
print("\n---IQR--- \n",IQR)
```

```python
data.shape
```

```python
#diabetes_dataset

data_out = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
data.shape,data_out.shape
```

```python
print(data[data['Diabetes'] == 'desired_value'])
```

```python
#diabetes_dataset

X=data_out.drop(columns=['Diabetes'])
y=data_out['Diabetes']
```

```python
train_X,test_X,train_y,test_y=train_test_split(X,y,test_size=0.3)
```

```python
train_X.shape,test_X.shape,train_y.shape,test_y.shape
```

```python
norm=MinMaxScaler().fit(train_X)
train_X_norm=norm.transform(train_X)
test_X_norm=norm.transform(test_X)
```

```python
def bmi_group(row):
    if row['BMI'] < 18.5:
        return '0'
    elif row['BMI'] > 18.5 and row['BMI'] < 25:
        return '1'
    elif row['BMI'] > 25 and row['BMI'] < 30:
        return '2'
    elif row['BMI'] > 30 and row ['BMI'] < 40:
        return '3'
    else:
        return '4'

data['bmi_status'] = data.apply(bmi_group,axis=1)

def age_group(row):
    if row['Age'] < 30:
        return '0'
    elif row['Age'] > 30 and row['Age'] > 62:
        return '1'
    else:
        return '2'

data['age_group'] = data.apply(age_group,axis=1)
```
